# 데이터 병합

#### 분리해서 분석했던 카테고리들을 병합해주기 

```python
olive_man=pd.concat([olive_skincare,olive_bodycare,olive_haircare,olive_makeup,olive_perfume,olive_sundrise])
olive_man
```

# price / pre_price 변수 

#### price / pre_price 의 분석을 위해 , 제거 

```python
olive_man['price']= olive_man['price'].str.replace(',', '').astype('int64')
olive_man['pre_price']=olive_man['pre_price'].str.replace(',',"").astype('int64') #할인상품이 아닌경우 할인전 가격은 0
```

```python
olive_man.head()
```

# brand_name encoded ( brand_ name 텍스트를 숫자로 )

```python
olive_man['brand_name'].values
```

```python
olive_man_cat=olive_man['brand_name']
olive_man_cat.head(10)
```

#### 각 카테고리를 다른 정수값으로 매핑해주는 factorize() 매서드 사용

```python
olive_man_cat_encoded,olive_man_cat_categories=olive_man_cat.factorize()
olive_man_cat_encoded
olive_man_cat_categories
```

```python
olive_man_cat_encoded[:10]
```

+ 텍스트가 숫자로 바뀐것 확인

#### 새로운 변수 brand_name2

```python
olive_man['brand_name2']=olive_man_cat_encoded
```

```python
olive_man.head()
```
